# Brownian Motion Simulations

Examples using `BrownianMotion` from `xpectral.data` and plotting paths with the `df.bokeh` accessor.

```python
%load_ext autoreload
%autoreload 2

# Other imports
from bokeh.io import output_notebook
from bokeh.io import show
from bokeh.palettes import Paired
import xpectral
from xpectral.charts.theme_manager import theme
from xpectral.data import BrownianMotion

output_notebook(hide_banner=True)
theme.set("ocean")

N_PATHS = 5
N_STEPS = 252
COLORS = Paired[10][:N_PATHS]
```

## Standard Brownian Motion

Cumulative Gaussian increments: $W_t = \sum_{i=1}^{t} \epsilon_i \sqrt{\Delta t}$

```python
bm = BrownianMotion(n_steps=N_STEPS, n_paths=N_PATHS, seed=42)
df_sbm = bm.standard()

path_cols = [c for c in df_sbm.columns if c.startswith("path_")]

fig = df_sbm.bokeh(
    title="Standard Brownian Motion",
    width=600,
    height=320,
    tools="pan,wheel_zoom,reset,save",
    x_axis_label="Step",
    y_axis_label="W(t)",
)
for col, color in zip(path_cols, COLORS):
    fig.line(x="step", y=col, line_width=1.5, line_color=color, line_alpha=0.85, legend_label=col)

fig.legend.location = "top_left"
fig.legend.click_policy = "hide"
show(fig)
```

## Geometric Brownian Motion

Exact solution: $S_t = S_0 \exp\!\left(\left(\mu - \tfrac{1}{2}\sigma^2\right)t + \sigma W_t\right)$

```python
bm = BrownianMotion(n_steps=N_STEPS, n_paths=N_PATHS, seed=42)
df_gbm = bm.geometric(mu=0.05, sigma=0.2, s0=100.0)

path_cols = [c for c in df_gbm.columns if c.startswith("path_")]

fig = df_gbm.bokeh(
    title="Geometric Brownian Motion  (μ=0.05, σ=0.20, S₀=100)",
    width=600,
    height=320,
    tools="pan,wheel_zoom,reset,save",
    x_axis_label="Step",
    y_axis_label="S(t)",
)
for col, color in zip(path_cols, COLORS):
    fig.line(x="step", y=col, line_width=1.5, line_color=color, line_alpha=0.85, legend_label=col)

fig.legend.location = "top_left"
fig.legend.click_policy = "hide"
show(fig)
```
